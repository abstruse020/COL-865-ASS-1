```python
# from pytorch_tabnet.tab_model import TabNetClassifier, TabNetRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import pandas as pd
import numpy as np
import arff
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%%time
traffic_db_path = '../Traffic_Violations/dataset.arff'
traffic_data = arff.load(open(traffic_db_path,'r'))

traffic_columns = [c[0] for c in traffic_data['attributes']]
print('columns',traffic_columns)

traffic_df = pd.DataFrame(traffic_data['data'], columns=traffic_columns)


traffic_df.head()
```

```python
''' details
Parse it-
date_of_stop - parse it into date
time_of_stop - parse it into time

Remove it-
agency -- just 1 value
description --  has sentences
model -- 19462  distinct values so whats the point
geolocation -- same as longitude and latitude

Fill None values --
location -- fill mean / new class
year -- 9647 missing attributes

Take None values as new class -
search_conducted -- 600828 missing attributes
search_disposition -- 1509055 missing attributes
search_outcome -- 619994 missing attributes
search_reason -- 1509055 missing attributes
search_reason_for_stop -- 600998 missing attributes
search_type -- 1509063 missing attributes
search_arrest_reason -- 1531303 missing attributes
state -- 59 missing attributes
make -- 9762 missing attributes
color -- 18657 missing attributes
article -- 76873 missing attributes
driver_city -- 387 missing attributes
driver_state -- 11 missing attributes
dl_state -- 929 missing attributes

Numeric Columns
date_of_stop, time_of_stop, latitude, longitude, year

Categorical Columns
subagency, location, accident, belts, personal_injury, property_damage, fatal, commercial_license, hazmat, commercial_vehicle, alcohol, work_zone, search_conducted, search_disposition, search_outcome, search_reason, search_reason_for_stop, search_type, search_arrest_reason, state, vehicletype, make, color, charge, article, contributed_to_accident, race, gender, driver_city, driver_state, dl_state, arrest_type


'''


remove_cols = ['agency','description','model',
'geolocation', 'location', 'driver_city']
fill_none_values_categorical = []
fill_none_values_numeric = ['year']
new_none_class = ['search_conducted', 'search_disposition', 'search_outcome', 'search_reason', 'search_reason_for_stop', 'search_type', 'search_arrest_reason', 'state', 'make', 'color', 'article', 'driver_state', 'dl_state']

categorical_cols = ['subagency', 'location', 'accident', 'belts', 'personal_injury', 'property_damage', 'fatal', 'commercial_license', 'hazmat', 'commercial_vehicle', 'alcohol', 'work_zone', 'search_conducted', 'search_disposition', 'search_outcome', 'search_reason', 'search_reason_for_stop', 'search_type', 'search_arrest_reason', 'state', 'vehicletype', 'make', 'color', 'charge', 'article', 'contributed_to_accident', 'race', 'gender', 'driver_state', 'dl_state', 'arrest_type']

numerical_cols = ['date_of_stop', 'time_of_stop', 'latitude', 'longitude', 'year']

all_missing_values_col = new_none_class + fill_none_values_categorical + fill_none_values_numeric
```

```python
## uniformly distributed across these labels so removing the 4 columns

max_rows = traffic_df.shape[0]
print('max rows',max_rows)
remove_cols
for col in traffic_columns:
    
    if traffic_df[col].isna().sum() > 1000 and col not in numerical_cols:
        print(col,'None Count:',traffic_df[col].isna().sum())
        print( traffic_df[traffic_df[col].notna()].groupby('violation_type')['seqid'].count())
        print('-'*20)
        remove_cols.append(col)
        
## removing cols
print('cols to remove:', remove_cols)
traffic_df = traffic_df.drop(columns=remove_cols)

## adding None as a seperate class label
for col in new_none_class:
    if col not in remove_cols:
        print('col:', col)
        traffic_df[col] = traffic_df[col].fillna(value='None_class')
        
## categorical to ohe
traffic_df_dummy = traffic_df.copy()

for col in categorical_cols:
    print('col:', col)
    if col not in remove_cols and len(traffic_df[col].unique()) < 100:
        print('\tconvert to ohe')
        traffic_df = pd.concat([traffic_df.drop(columns=col),pd.get_dummies(traffic_df[col], prefix=col)], axis=1)
    elif col not in remove_cols and len(traffic_df[col].unique()) >= 100:
        print('\tdrop')
        traffic_df = traffic_df.drop(columns = col)
        remove_cols.append(col)
        
# fill none values  in numeric (i.e) year
# check all numeric are actually numeric already fixed
#parse date and time

for col in fill_none_values_numeric:
    traffic_df[col] = traffic_df[col].fillna(value=traffic_df[col].mean())

traffic_df['date_of_stop'] = pd.to_datetime(traffic_df['date_of_stop']).astype(int)/10e9
traffic_df['time_of_stop'] = pd.to_timedelta(traffic_df_dummy['time_of_stop']).astype(int)/10e9
```

```python
## making duplicate row if dataset only has 1 row for given class
label_grp_df = traffic_df.groupby('violation_type')
min_label_size = 2
test_size = 0.2
small_sized_labels = []
small_sized_labels_df = pd.DataFrame(columns=traffic_df.columns)
# y_df = pd.get_dummies(traffic_df.iloc[:,-1])
train_df = pd.DataFrame(columns=traffic_df.columns)
test_df = pd.DataFrame(columns=traffic_df.columns)

for grp_label, grp in label_grp_df:
    grp_size = len(grp)
    print(grp_label, grp_size)
    if grp_size < min_label_size:
        small_sized_labels.append(grp_label)
        #print(f'grp:{grp}')
        for row in grp.iterrows():
            #print('row',row[1])
            small_sized_labels_df = small_sized_labels_df.append(row[1], ignore_index=True)
            train_df = train_df.append(row[1], ignore_index=True)
            test_df = test_df.append(row[1], ignore_index=True)
    else:
        temp_train, temp_test = train_test_split(grp, test_size=0.2)
        print(f'\t temp_train size:{temp_train.shape}, temp_test size:{temp_test.shape}')
        train_df = pd.concat([train_df, temp_train], ignore_index=True)
        test_df = pd.concat([test_df, temp_test], ignore_index=True)
        
print(f'\nsmall sized labels:{small_sized_labels}')
train_df = train_df.sample(frac=1)
test_df = test_df.sample(frac=1)
small_sized_labels_df
```

```python
X_train = train_df.drop(columns=['seqid', 'violation_type'])
y_train_non_ohe = train_df['violation_type']
y_train = pd.get_dummies(train_df['violation_type'])
X_test = test_df.drop(columns=['seqid', 'violation_type'])
y_test_non_ohe = test_df['violation_type']
y_test = pd.get_dummies(test_df['violation_type'])
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
%%time
from sklearn.ensemble import BaggingRegressor, BaggingClassifier
from sklearn.model_selection import cross_val_score

from scikit_obliquetree.HHCART import HouseHolderCART, HHCARTNode
from scikit_obliquetree.segmentor import MSE, MeanSegmentor, ABCMeta

# clf = BaggingClassifier(
#         HHCARTNode(ABCMeta(), MeanSegmentor(), max_depth=3),
#         n_estimators=100,
#         n_jobs=-1
# )

reg = BaggingRegressor(
    HouseHolderCART(MSE(), MeanSegmentor(), max_depth=20),
    n_estimators=50,
    n_jobs=-1,
)
```

```python
%%time
reg.fit(X_train, np.argmax(y_train.values, axis=1))
```

```python
%%time
y_train_pred = reg.predict(X_train)
accuracy_score(np.argmax(y_train.values, axis=1), np.round(y_train_pred).astype('int'))
```

```python
%%time
y_test_pred = reg.predict(X_test)
accuracy_score(np.argmax(y_test.values, axis=1), np.round(y_test_pred).astype('int'))
```
